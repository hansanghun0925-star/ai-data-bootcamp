# 문제 1. 기초통계량 계산

**📘 문제**

- 온라인 쇼핑몰 고객 1,000명을 대상으로 1점부터 10점 사이의 만족도 조사를 실시했습니다.
  이 데이터를 바탕으로 고객 만족도의 **중심 경향성**과 **분포 특성**을 파악해 봅시다.

```python
# 데이터 생성
np.random.seed(2025)
satisfaction = np.random.normal(7.5, 1.5, 1000)
satisfaction = np.clip(satisfaction, 1, 10)
df1 = pd.DataFrame({'score': satisfaction})
```

**📌 아래를 수행해 보세요:**

- 평균, 중앙값, 최빈값, 표준편차를 계산해 봅시다.
- 히스토그램을 그리고 평균과 중앙값 선으로 표시해 봅시다.
- 데이터의 분포 형태를 해석한 후, 마케팅 전략 수립에 어떻게 활용할 수 있을지 생각해 봅시다.

```python
# [문제 1] 기초통계량 계산
# 기초통계량 계산
mean_score = df1['score'].mean()
median_score = df1['score'].median()
mode_score = df1['score'].round(1).mode()[0]  # 연속형이라 소수1자리로 반올림 후 최빈값
std_score = df1['score'].std()

print(f"평균     : {mean_score:.3f}")
print(f"중앙값   : {median_score:.3f}")
print(f"최빈값   : {mode_score:.1f}")
print(f"표준편차 : {std_score:.3f}")

# 히스토그램 + 평균/중앙값 선
plt.figure(figsize=(9, 5))
plt.hist(df1['score'], bins=30, color='skyblue', edgecolor='white', alpha=0.8)
plt.axvline(mean_score, color='red', linestyle='--', linewidth=2, label=f'평균 {mean_score:.2f}')
plt.axvline(median_score, color='green', linestyle='-', linewidth=2, label=f'중앙값 {median_score:.2f}')
plt.title('고객 만족도 분포')
plt.xlabel('만족도 점수')
plt.ylabel('빈도')
plt.legend()
plt.show()
```

# 문제 2. 평균 vs 중앙값 비교

**📘 문제**

- A지역과 B지역의 연봉 데이터를 비교해 봅시다.
- B 지역에는 일부 고소득자가 포함되어 있습니다.

```python
# 데이터 생성
np.random.seed(2025)

# A 지역: 평균 300, 표준편차 50
region_a = np.random.normal(300, 50, 100)

# B 지역: 일반 소득자 + 극단적인 고소득자 포함
region_b = np.concatenate([
    np.random.normal(280, 40, 95),    # 일반 소득자 95명
    np.random.normal(1500, 200, 5)    # 고소득자 5명
])

df2 = pd.DataFrame({
    'region': ['A'] * 100 + ['B'] * 100,
    'salary': np.concatenate([region_a, region_b])
})
```

- 지역별 박스플롯(Boxplot)을 그려서 분포의 차이를 시각적으로 확인해 봅시다.

```python
# [문제 2] 평균 vs 중앙값 비교
# 지역별 평균/중앙값
summary2 = df2.groupby('region')['salary'].agg(['mean', 'median'])
print(summary2)

# 지역별 박스플롯
plt.figure(figsize=(8, 5))
sns.boxplot(data=df2, x='region', y='salary')
plt.title('지역별 연봉 분포')
plt.xlabel('지역')
plt.ylabel('연봉')
plt.show()
```

# 문제 3. 이상값과 분산 비교

**📘 문제**

- 1반과 2반의 시험 점수 분포를 비교해 봅시다.
- 2반에는 극단적인 이상값(예: 부정행위로 0점)이 포함되어 있습니다.

```python
# 데이터 생성
np.random.seed(2025)  # 재현 가능성을 위해 시드 고정

# 1반: 평균 75, 표준편차 8
class1 = np.random.normal(75, 8, 30)

# 2반: 일반 학생 + 이상값(0점, 130점)
class2 = np.concatenate([
    np.random.normal(75, 15, 28),  # 일반 학생
    [0, 130]                       # 극단적인 이상값
])

df3 = pd.DataFrame({
    'class': ['1반'] * 30 + ['2반'] * 30,
    'score': np.concatenate([class1, class2])
})
```

**📌 아래를 수행해 보세요:**

1. 각 반의 평균, 표준편차, 최소/최대값을 계산해 봅시다.

3. 이상값이 통계 지표에 어떤 영향을 주는지 생각해 봅시다.

```python
# [문제 3] 이상값과 분산 비교
# 반별 평균/표준편차/최소/최대
summary3 = df3.groupby('class')['score'].agg(['mean', 'std', 'min', 'max'])
print(summary3)

# 반별 박스플롯
plt.figure(figsize=(8, 5))
sns.boxplot(data=df3, x='class', y='score')
plt.title('반별 시험 점수 분포')
plt.xlabel('반')
plt.ylabel('점수')
plt.show()
```

# 문제 4. IQR 이상값 제거

**📘 문제**

- 앞의 문제에서 2반에는 **극단적인 이상값(예: 매우 낮은 점수)**이 포함되어 있어
  통계 지표, 특히 평균에 큰 영향을 주는 것을 확인했습니다.
- 이번에는 **IQR(사분위 범위)**을 기준으로 이상값을 제거한 뒤, 이상값 제거 전후의 평균을 비교해 봅시다

<br>

**📌 아래를 수행해 보세요:**

- 2반의 점수 데이터를 대상으로 IQR 기준을 사용해 이상값을 판별하고 제거해 봅시다.
- 이상값 제거 전과 제거 후의 평균을 각각 계산해 봅시다.
- 이상값 제거가 평균에 어떤 영향을 주는지 설명해 봅시다.
- 필요하다면 중앙값과도 비교해 봅시다.

```python
# [문제 4] IQR 이상값 제거
# 2반 점수만 추출
class2 = df3[df3['class'] == '2반']['score']

# IQR 계산
Q1 = class2.quantile(0.25)
Q3 = class2.quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
print(f"Q1={Q1:.2f}, Q3={Q3:.2f}, IQR={IQR:.2f}")
print(f"정상 범위: {lower:.2f} ~ {upper:.2f}")

# 이상값 제거
class2_clean = class2[(class2 >= lower) & (class2 <= upper)]
removed = class2[(class2 < lower) | (class2 > upper)]
print(f"제거된 이상값: {sorted(removed.tolist())}")

# 제거 전후 비교
print(f"\n제거 전 평균 : {class2.mean():.2f}")
print(f"제거 후 평균 : {class2_clean.mean():.2f}")
print(f"중앙값       : {class2.median():.2f}")
```

1. 이상값 제거 전후 평균은 어떻게 달라졌나요?
   - 평균이 중앙값에 가까워졌나요?
   - 이상값 제거가 평균에 어떤 방향으로 영향을 주었나요?

2. 왜 IQR을 기준으로 이상값을 판단할까요?
   - IQR은 데이터를 어떻게 나누고, 어떤 방식으로 이상값을 정의하나요?
   - 평균이나 표준편차 기준과 비교했을 때 어떤 점이 더 강건할까요?

3. 이상값은 언제 제거하고, 언제 유지해야 할까요?
   - 분석 목적에 따라 이상값은 중요한 정보가 될 수 있습니다.

4. 중앙값과 평균은 각각 어떤 상황에서 더 유용한 지표인가요?
   - 이상값이 있을 때 어떤 지표가 더 신뢰할 수 있나요?
   - 두 지표를 함께 보면 어떤 장점이 있을까요?

```python
# [문제 4] 해석
# - 이상값(0, 130) 제거 후 평균이 중앙값(≈73)에 더 가까워짐 -> 데이터의 실제 중심을 잘 반영.
# - IQR 기준(Q1-1.5*IQR ~ Q3+1.5*IQR)은 평균/표준편차를 안 쓰기 때문에 극단값의 영향을
#   덜 받아 더 강건(robust)함.
# - 이상값은 항상 지우는 게 아니라 목적에 따라 판단: 부정행위(0점)는 제거 타당하지만,
#   진짜 고득점(130 같은 실제값)이라면 중요한 정보일 수 있음.
# - 이상값이 있을 땐 중앙값이 더 신뢰할 만하고, 평균과 함께 보면 왜곡 여부를 파악하기 좋음.
```

# 문제 5. 변수 유형 분류

**📘 문제**

# 문제 9. 정규분포와 표준편차

**📘 문제**

- 정규분포와 표준편차의 개념을 실제 데이터에 적용해 보겠습니다.
- 정규분포를 따르는 점수 데이터에서 다음을 확인해 봅시다:
  - 평균 ±1σ, 평균 ±2σ 범위에 속한 값들의 비율은 실제로 얼마나 되는가?
  - 이 결과가 이론적인 68-95-99 법칙과 얼마나 유사한지 비교해 봅시다.

<br>

**📌 아래를 수행해 보세요:**

1. 평균이 70, 표준편차가 10인 정규분포 데이터를 1000개 생성해 봅시다.
2. 평균을 중심으로 ±1σ, ±2σ 범위에 포함되는 데이터의 비율을 계산해 봅시다.
3. 결과를 바탕으로 68-95-99 법칙과 비교해 봅시다.

```python
# [문제 9] 정규분포와 표준편차
# 1. 정규분포 데이터 생성
np.random.seed(2025)
data = np.random.normal(70, 10, 1000)
mean_d = data.mean()
std_d = data.std()

# 2. ±1σ, ±2σ 범위 비율 계산
within_1 = np.mean((data >= mean_d - std_d) & (data <= mean_d + std_d)) * 100
within_2 = np.mean((data >= mean_d - 2*std_d) & (data <= mean_d + 2*std_d)) * 100

print(f"평균 {mean_d:.2f}, 표준편차 {std_d:.2f}")
print(f"±1σ 범위 비율 : {within_1:.1f}%  (이론 68%)")
print(f"±2σ 범위 비율 : {within_2:.1f}%  (이론 95%)")

# 시각화
plt.figure(figsize=(9, 5))
plt.hist(data, bins=30, color='lightblue', edgecolor='white')
for k, c in [(1, 'green'), (2, 'orange')]:
    plt.axvline(mean_d - k*std_d, color=c, linestyle='--', label=f'±{k}σ')
    plt.axvline(mean_d + k*std_d, color=c, linestyle='--')
plt.axvline(mean_d, color='red', linewidth=2, label='평균')
plt.title('정규분포와 표준편차 구간')
plt.xlabel('값'); plt.ylabel('빈도'); plt.legend()
plt.show()
```

3. 표준편차는 왜 중요한 지표인가요?
4. 이 개념이 실제로 어떻게 활용될 수 있을까요?

```python
# [문제 9] 해석
# - ±1σ에 약 68%, ±2σ에 약 95%가 들어와 68-95-99 법칙과 거의 일치.
# - 표본이 1000개로 충분히 커서 이론값에 잘 수렴함.
# - 표준편차는 "데이터가 평균에서 얼마나 퍼져 있는지"를 나타내는 핵심 지표.
# - 활용: 이상치 탐지(±2σ, ±3σ 벗어난 값), 품질관리(관리 한계선), 시험 성적의
#   상대적 위치(표준점수, z-score) 계산 등.
```

# 문제 11. 단순 선형 회귀 분석

**📘 문제**

- 공부 시간과 성적 간의 관계를 회귀 분석으로 확인해봅시다.

```python
# 데이터 생성
from sklearn.linear_model import LinearRegression

np.random.seed(505)
study_time = np.random.normal(5, 2, 100)
score = 50 + study_time * 8 + np.random.normal(0, 5, 100)
df11 = pd.DataFrame({'study_time': study_time, 'score': score})
```

**📌 아래를 수행해 보세요:**

1. 공부 시간(`study_time`)과 성적(`score`)간의 관계를 단순 선형 회귀모델로 분석합니다.
2. 회귀 모델을 학습하고, 기울기(계수)와 절편을 출력해 봅시다.
3. 산점도 위에 회귀선을 시각화해 봅시다.

```python
# [문제 11] 단순 선형 회귀 분석
# 1. 학습 데이터 준비
X = df11[['study_time']]   # 2차원 형태
y = df11['score']

# 2. 회귀 모델 학습
model = LinearRegression()
model.fit(X, y)

slope = model.coef_[0]
intercept = model.intercept_
r2 = model.score(X, y)
print(f"기울기(계수) : {slope:.3f}")
print(f"절편         : {intercept:.3f}")
print(f"R^2          : {r2:.3f}")
print(f"회귀식: score = {slope:.2f} * study_time + {intercept:.2f}")

# 3. 산점도 + 회귀선
plt.figure(figsize=(9, 5))
plt.scatter(df11['study_time'], df11['score'], alpha=0.6, label='실제 데이터')
x_line = np.linspace(df11['study_time'].min(), df11['study_time'].max(), 100)
y_line = model.predict(x_line.reshape(-1, 1))
plt.plot(x_line, y_line, color='red', linewidth=2, label='회귀선')
plt.title('공부 시간과 성적의 관계')
plt.xlabel('공부 시간'); plt.ylabel('성적'); plt.legend()
plt.show()
```

# 문제 12. A/B 테스트 해석

**📘 문제**

- 두 그룹의 전환율 차이를 계산하고 해석해봅시다.

```python
# 데이터 생성
np.random.seed(606)
group_a = np.random.binomial(1, 0.12, 1000)
group_b = np.random.binomial(1, 0.15, 1000)
```

**📌 아래를 수행해 보세요:**

1. A 그룹과 B 그룹의 전환 데이터를 살펴봅니다.
2. 각 그룹의 전환율을 계산하고, 차이를 확인합니다.
3. 이 차이가 우연인지, 의미 있는 차이인지 생각해 봅시다.

```python
# [문제 12] A/B 테스트 해석
from scipy.stats import chi2_contingency

# 1. 각 그룹 전환율
rate_a = group_a.mean()
rate_b = group_b.mean()
print(f"A 그룹 전환율 : {rate_a:.3f} ({group_a.sum()}/{len(group_a)})")
print(f"B 그룹 전환율 : {rate_b:.3f} ({group_b.sum()}/{len(group_b)})")
print(f"전환율 차이   : {rate_b - rate_a:.3f}")

# 2. 차이가 통계적으로 유의한지 검정 (카이제곱 검정)
table = np.array([
    [group_a.sum(), len(group_a) - group_a.sum()],
    [group_b.sum(), len(group_b) - group_b.sum()]
])
chi2, p_value, dof, expected = chi2_contingency(table)
print(f"\n카이제곱 통계량 : {chi2:.3f}")
print(f"p-value        : {p_value:.4f}")
print("=> 유의미한 차이" if p_value < 0.05 else "=> 통계적으로 유의하지 않음(우연일 수 있음)")

# 3. 전환율 시각화
plt.figure(figsize=(7, 5))
sns.barplot(x=['A', 'B'], y=[rate_a, rate_b])
plt.title('A/B 그룹 전환율 비교')
plt.ylabel('전환율')
plt.show()
```
