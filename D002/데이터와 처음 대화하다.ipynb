````python
# ─────────────────────────────────────────────
# 모두마켓 데이터 생성 — 이 셀 하나로 오늘 쓸 데이터가 모두 준비됩니다.
# (실제 현장처럼 '오염'을 일부러 심어 둡니다: 결측·이상치·표기 혼재·날짜 포맷 혼재·중복)
# ─────────────────────────────────────────────
np.random.seed(42)

# 1) 고객(customers)
n_customers = 300
customers = pd.DataFrame({
    "customer_id": [f"C{str(i).zfill(4)}" for i in range(1, n_customers + 1)],
    "age": np.random.normal(35, 9, n_customers).round().astype(int),
    "gender": np.random.choice(["M", "F"], n_customers),
    "region": np.random.choice(["서울", "경기", "부산", "인천", "대구"], n_customers),
    "membership": np.random.choice(["basic", "premium", "vip"], n_customers, p=[0.6, 0.3, 0.1]),
})
# 오염 심기: 나이 이상치, 결측, 지역/멤버십 표기 혼재
customers.loc[5, "age"] = 999          # 입력 실수로 보이는 이상치
customers.loc[10, "age"] = -3          # 음수 나이(불가능한 값)
customers.loc[[20, 21, 22], "gender"] = np.nan
customers.loc[30, "region"] = " 서울 "  # 앞뒤 공백
customers.loc[31, "region"] = "Seoul"   # 영문 표기 혼재
customers.loc[40, "membership"] = "VIP"  # 대소문자 혼재

# 2) 상품(products)
categories = ["패션", "뷰티", "식품", "가전", "도서"]
n_products = 40
products = pd.DataFrame({
    "product_id": [f"P{str(i).zfill(3)}" for i in range(1, n_products + 1)],
    "category": np.random.choice(categories, n_products),
    "price": np.random.choice([9900, 19900, 29900, 49900, 89900, 129900], n_products),
})

# 3) 주문(orders)
n_orders = 2000
order_customer = np.random.choice(customers["customer_id"], n_orders)
order_product = np.random.choice(products["product_id"], n_orders)
price_map = products.set_index("product_id")["price"]
quantity = np.random.choice([1, 1, 1, 2, 2, 3], n_orders)
amount = price_map.loc[order_product].values * quantity

orders = pd.DataFrame({
    "order_id": [f"O{str(i).zfill(5)}" for i in range(1, n_orders + 1)],
    "customer_id": order_customer,
    "product_id": order_product,
    "quantity": quantity,
    "amount": amount.astype(float),
    "channel": np.random.choice(["web", "app", "app ", "APP"], n_orders, p=[0.45, 0.45, 0.05, 0.05]),
})

# 날짜 포맷 혼재(문자열로 저장 — 일부러 통일하지 않음)
base_dates = pd.to_datetime("2025-01-01") + pd.to_timedelta(np.random.randint(0, 120, n_orders), unit="D")
date_strings = []
for i, d in enumerate(base_dates):
    if i % 3 == 0:
        date_strings.append(d.strftime("%Y-%m-%d"))
    elif i % 3 == 1:
        date_strings.append(d.strftime("%Y/%m/%d"))
    else:
        date_strings.append(d.strftime("%Y%m%d"))
orders["order_date"] = date_strings

# 오염 심기: 금액 결측, 수량 이상치, 중복 행
orders.loc[np.random.choice(n_orders, 35, replace=False), "amount"] = np.nan
orders.loc[7, "quantity"] = 100        # 비정상적으로 큰 주문 수량
orders = pd.concat([orders, orders.iloc[[0, 1]]], ignore_index=True)  # 중복 주문 2건

print("모두마켓 데이터 생성 완료")
print("customers:", customers.shape, "| products:", products.shape, "| orders:", orders.shape)
````

## 자료형과 컬렉션 — 데이터 한 칸, 한 줄을 무엇에 담을까

# 1. 자료형과 컬렉션 — 데이터 한 칸, 한 줄을 무엇에 담을까

모두마켓 데이터를 보면 고객의 나이(`35`), 지역(`"서울"`), 회원 여부(`True`/`False`) 같은 값이 칸칸이 들어 있습니다. 컴퓨터는 이 값들을 **종류별로 다르게** 취급합니다. `35`는 계산할 수 있는 숫자이고, `"서울"`은 글자입니다. 이 종류를 **자료형(data type)** 이라고 합니다.

> ❓ **이 파트에서 답할 질문:** 데이터 한 칸은 무엇으로 담고, 여러 칸을 묶을 때는 무엇으로 담을까요?

## 💡 쉽게 말하면 — 정리함에 비유하기

자료형과 컬렉션을 정리함에 비유해봅시다.

```text
자료형(한 칸)        컬렉션(여러 칸을 묶는 그릇)
  35  (정수)          [35, 28, 41]      ← 리스트: 순서가 있고 바꿀 수 있는 줄
 "서울" (문자열)       (127.0, 37.5)     ← 튜플: 순서가 있지만 못 바꾸는 줄(좌표 등)
 True (불리언)        {"서울","부산"}     ← 집합: 중복 없는 모음
 3.14 (실수)          {"name":"홍길동"}   ← 딕셔너리: 이름표(key)로 값을 찾는 사물함
```

값 하나를 담는 것이 **자료형**, 값 여럿을 한 그릇에 모으는 것이 **컬렉션(collection)** 입니다.

## 자세히 알아보기

먼저 값 하나를 담는 기본 자료형입니다.

| 자료형 | 영어 | 예시 | 데이터에서의 의미 |
| --- | --- | --- | --- |
| 정수 | int | `35`, `2000` | 나이, 주문 건수처럼 셀 수 있는 수 |
| 실수 | float | `3.14`, `19900.0` | 평균 금액, 비율처럼 소수가 가능한 수 |
| 문자열 | str | `"서울"`, `"web"` | 지역, 채널처럼 글자로 된 값 |
| 불리언 | bool | `True`, `False` | 조건의 참/거짓 (프리미엄 회원인가?) |

다음은 값 여럿을 묶는 컬렉션 4종입니다. 이 차이가 실무에서 자주 헷갈리니 표로 정리합니다.

| 컬렉션 | 영어 | 표기 | 순서 | 수정 | 중복 | 주 용도 |
| --- | --- | --- | --- | --- | --- | --- |
| 리스트 | list | `[ ]` | O | O | O | 값들을 순서대로 쌓을 때 |
| 튜플 | tuple | `( )` | O | X | O | 바뀌면 안 되는 묶음(좌표, 설정) |
| 집합 | set | `{ }` | X | O | X | 중복 제거, "어떤 종류가 있나" |
| 딕셔너리 | dict | `{key:value}` | O(3.7+) | O | key 중복X | 이름표로 값을 찾을 때 |

> **읽는 법:** 집합 `regions_seen`에 `"서울"`을 두 번 넣었지만 출력에는 한 번만 나옵니다. 이 성질 덕분에 "이 컬럼에 어떤 종류의 값이 있나?"를 알아낼 때 집합을 씁니다. 딕셔너리는 `customer["region"]`처럼 **이름표(key)** 로 값을 꺼냅니다.

## 함수와 클래스 — 반복을 묶는 법

# 2. 함수와 클래스 — 반복을 묶는 법

Part 1에서 주문 금액 리스트의 합을 `sum()`으로 구했습니다. 그런데 "평균을 구하고, 부가세를 붙이고, 천 단위 콤마를 찍어라" 같은 작업을 데이터가 바뀔 때마다 매번 복사-붙여넣기 한다면 어떨까요? 실수가 늘고 코드가 지저분해집니다.

> ❓ **이 파트에서 답할 질문:** 같은 작업을 반복하지 않으려면 어떻게 한 덩어리로 묶을 수 있을까요?

## 💡 쉽게 말하면 — 함수는 "자판기"

함수(function)는 **자판기**와 같습니다. 정해진 입력(돈, 버튼)을 넣으면 정해진 출력(음료)이 나옵니다. 한 번 만들어두면 누구나, 몇 번이든 같은 방식으로 씁니다.

```text
입력(input)  →  [ 함수: 무엇을 할지 정의 ]  →  출력(output)
 금액, 수량        주문 총액을 계산한다          총액
```

## 자세히 알아보기

```python
def 함수이름(입력1, 입력2):
    결과 = 입력1 * 입력2
    return 결과          # return으로 결과를 돌려줍니다
```

- `def` : 함수를 정의하겠다는 신호
- `return` : 계산 결과를 함수 밖으로 돌려주는 명령 (없으면 `None`이 나옵니다)
- **매개변수(parameter)** : 함수가 받는 입력의 이름

함수의 가장 큰 가치는 **재사용**과 **이름**입니다. `calc_total`이라는 이름을 보면 코드를 읽지 않아도 무엇을 하는지 짐작됩니다.

> **읽는 법:** `vat_rate=0.1`처럼 기본값이 있는 매개변수는 생략하면 자동으로 그 값을 씁니다. `lambda x: ...`는 `def` 없이 만드는 짧은 함수로, "각 값에 똑같은 변환을 적용"하는 상황(Part 4의 pandas)에서 다시 만납니다.

## 💡 쉽게 말하면 — 클래스는 "붕어빵 틀"

함수가 하나의 작업이라면, **클래스(class)** 는 *관련된 데이터와 작업을 하나로 묶은 설계도* 입니다. 붕어빵 틀(클래스)로 붕어빵(객체)을 여러 개 찍어내는 것에 비유합니다.

사실 우리가 쓰는 `pandas.DataFrame`도 하나의 클래스입니다. `df.describe()`, `df.head()`처럼 점(`.`)을 찍고 쓰는 기능들이 바로 클래스에 묶인 작업(메서드)이에요. 그래서 클래스를 알면 라이브러리 사용법이 한결 또렷해집니다.

## 데이터로 확인해 봅시다

- 분석가는 클래스를 **직접 설계하는 일은 드물지만**, 라이브러리가 준 클래스를 *사용*합니다. `df.head()`의 `df`가 객체, `.head()`가 메서드입니다. 오늘 이 구조를 알아두면 Part 4부터 pandas가 훨씬 친숙해집니다.
- 반대로 **함수**는 분석가가 매일 만듭니다. "결측치 비율을 계산하는 함수", "품질 리포트를 출력하는 함수"처럼요. 이 모듈의 마지막 날(마무리)에는 여러분이 직접 **품질 리포트 함수**를 작성하게 됩니다.

> 📌 **다른 산업에서는?** 재사용 가능한 함수는 모든 산업의 분석 코드를 지탱합니다. 금융의 리스크 점수 계산, 제조의 불량률 집계, 마케팅의 전환율 산출 — 모두 한 번 잘 만든 함수를 반복 호출하는 구조입니다.

````python
# 예제: 주문 총액(부가세 포함)을 계산하는 함수
def calc_total(price, quantity, vat_rate=0.1):
    # 가격 x 수량에 부가세를 더한 총액을 반환합니다.
    subtotal = price * quantity
    total = subtotal * (1 + vat_rate)
    return total

# 한 번 정의했으니 몇 번이든 재사용합니다.
print("19900원 2개 총액:", calc_total(19900, 2))
print("49900원 1개 총액:", calc_total(49900, 1))
print("9900원 3개 총액 :", calc_total(9900, 3))
````

## NumPy 배열 — 숫자 덩어리를 한 번에

# 3. NumPy 배열 — 숫자 덩어리를 한 번에

모두마켓의 주문은 2,000건, 실제 회사라면 수백만 건입니다. 모든 주문 금액에 10% 부가세를 더하려면 어떻게 할까요? Part 1~2의 방식이라면 리스트를 하나씩 돌며 곱해야 합니다. 데이터가 커질수록 느려지고 코드도 길어집니다.

> ❓ **이 파트에서 답할 질문:** 수많은 숫자에 같은 계산을 한 번에, 그것도 빠르게 적용하려면?

## 💡 쉽게 말하면 — "전체에 동시에" (벡터화)

NumPy(넘파이)의 핵심 아이디어는 **벡터화(vectorization)** 입니다. 반복문 없이 배열 전체에 연산을 한 번에 거는 것이죠.

```text
[일반 리스트]  19900 → x1.1   (한 칸씩 돌며 곱하기 → 느림)
               29900 → x1.1
                9900 → x1.1

[NumPy 배열]  [19900, 29900, 9900] * 1.1   (전체에 동시에 → 빠름)
```

## 자세히 알아보기

- **ndarray** : NumPy의 배열. 같은 자료형의 숫자를 한 줄(또는 격자)로 담는 그릇.
- **벡터화(vectorization)** : 배열 전체에 사칙연산을 한 번에 적용. `arr * 1.1`처럼.
- **브로드캐스팅(broadcasting)** : 크기가 다른 배열·숫자끼리도 규칙에 따라 자동으로 맞춰 계산.
- **집계 함수** : `arr.mean()`, `arr.sum()`, `arr.std()`처럼 배열 전체를 한 값으로 요약.

> **읽는 법:** 같은 `* 2`인데 리스트는 내용이 **두 번 반복**되고, 배열은 **각 값이 2배**가 됩니다. 데이터 계산에서 우리가 원하는 건 거의 항상 후자입니다. 이것이 분석에서 리스트 대신 배열(그리고 pandas)을 쓰는 이유입니다.

> **읽는 법:** 같은 결과인데 벡터화가 훨씬 빠릅니다. NumPy가 내부적으로 최적화된 코드(C 언어)로 한꺼번에 처리하기 때문입니다. 데이터가 클수록 이 차이는 더 벌어집니다. (대용량 처리는 이 모듈 후반 Polars 편에서 더 깊이 다룹니다.)

## 데이터로 확인해 봅시다

- pandas의 컬럼은 **속을 들여다보면 NumPy 배열**입니다. 그래서 `df["amount"] * 1.1` 같은 코드가 자연스럽게 동작합니다. NumPy를 알면 pandas의 빠른 계산이 *왜* 빠른지 이해됩니다.
- `mean`, `std`, `max` 같은 집계 함수는 Part 5의 `describe()`가 내부에서 쓰는 바로 그 계산들입니다.

> 📌 **다른 산업에서는?** 벡터화 연산은 이미지(제조 품질 검사), 신호(헬스케어 심전도), 가격 시계열(금융) 등 "숫자가 아주 많은" 모든 분야의 계산을 떠받칩니다.

> 💡 **더 알고 싶다면 (선택):** NumPy 배열은 2차원(행렬)도 됩니다. `np.array([[1, 2], [3, 4]])`처럼요. 이때 `.sum(axis=0)`은 **열 방향**(세로로 더하기), `.sum(axis=1)`은 **행 방향**(가로로 더하기) 합입니다. pandas의 `df.sum(axis=0)`이 컬럼별 합인 것과 똑같은 규칙이에요. 지금 당장 외울 필요는 없지만, 앞으로 '축(axis)'이라는 말이 나오면 "어느 방향으로 계산하라는 거지?"를 묻는 신호라고 기억해두세요.

# 4. pandas DataFrame — 표 형태의 데이터 다루기

NumPy 배열은 **한 가지 자료형의 숫자**를 빠르게 다룹니다. 하지만 모두마켓 고객 데이터에는 나이(숫자), 지역(문자), 회원등급(문자)이 한 줄에 함께 있습니다. 이렇게 **여러 자료형이 섞인 표**를 다루려면 새 도구가 필요합니다.

> ❓ **이 파트에서 답할 질문:** 행과 열로 된 표(엑셀 시트 같은) 데이터를 어떻게 골라내고, 거르고, 계산할까요?

## 💡 쉽게 말하면 — "스마트한 엑셀 시트"

pandas(판다스)의 **DataFrame**은 엑셀 시트라고 생각하면 쉽습니다. 단, 코드로 다루기 때문에 훨씬 빠르고 정확하며 기록이 남습니다.


- **Series(시리즈)** : 한 개의 열. (인덱스를 가진 1차원 — 사실상 NumPy 배열 + 이름표)

## 자세히 알아보기 — NumPy 배열 vs pandas DataFrame

언제 무엇을 쓰는지는 이 표 하나로 정리됩니다.

| 구분 | NumPy 배열(ndarray) | pandas DataFrame |
| --- | --- | --- |
| 차원 | 주로 1·2차원 숫자 격자 | 행×열의 표 |
| 자료형 | **한 종류**(전부 숫자 등) | 열마다 **다른 자료형** 허용 |
| 이름표 | 없음(위치로 접근) | 행 인덱스·열 이름으로 접근 |
| 강점 | 순수 수치 계산이 가장 빠름 | 라벨·결측·혼합형 등 **실데이터** 처리 |
| 주 용도 | 행렬 연산, 모델 내부 계산 | 데이터 정제·탐색·집계(분석가의 주 무대) |

이제 진짜 모두마켓 데이터에 같은 기술을 써봅시다.

## 구조·품질 진단 — 분석 전에 데이터를 진찰하기

# 5. 구조·품질 진단 — 분석 전에 데이터를 진찰하기

Part 4 마지막에서 `.mean()`이 결측치를 조용히 빼고 평균을 냈습니다. 만약 절반이 비어 있었다면? 그 평균은 믿기 어렵습니다. 그래서 좋은 분석가는 계산보다 먼저 **데이터를 진찰**합니다. 의사가 처방 전에 진단부터 하듯이요.

> ❓ **이 파트에서 답할 질문:** 처음 받은 데이터의 "구조"와 "품질"을 어떤 순서로 확인해야 할까요?

## 💡 쉽게 말하면 — 건강검진 3종 세트

```text
1) shape    → "크기 재기"   몇 행 몇 열인가? (키·몸무게)
2) info     → "골격 보기"   각 열의 자료형과 결측은? (X-ray)
3) describe → "수치 검사"   숫자 열의 평균·범위·이상치 신호? (혈액검사)
```

이 세 가지를 **항상 이 순서로** 먼저 합니다. 여기에 결측(`isnull`)·중복(`duplicated`)·고윳값(`unique`) 확인을 더하면 기초 진단이 완성됩니다.

## 자세히 알아보기

| 도구 | 무엇을 알려주나 | 대표 사용 |
| --- | --- | --- |
| `df.shape` | (행 수, 열 수) | 데이터 규모 파악 |
| `df.info()` | 열별 자료형 + 결측이 아닌 개수 | 자료형·결측 한눈에 |
| `df.describe()` | 숫자 열의 개수·평균·표준편차·min·max·사분위 | 분포·이상치 신호 |
| `df.isnull().sum()` | 열별 결측치 개수 | 어디가 비었나 |
| `df.duplicated().sum()` | 중복 행 개수 | 중복 주문 탐지 |
| `df["col"].unique()` | 그 열에 등장한 값의 종류 | 표기 혼재 발견 |

> **읽는 법:** `info()`의 핵심은 두 가지입니다. (1) `Dtype` 열에서 `order_date`가 `object`(문자열)로 잡혀 있습니다 — 날짜인데 숫자로 못 쓰는 상태죠. (2) `Non-Null Count`에서 `amount`가 전체보다 작다면 그만큼 **결측**이 있다는 뜻입니다.

> **읽는 법:** `describe()`에서 이상 신호를 찾습니다. `customers`의 `age`에서 `max`가 **999**, `min`이 **음수**라면 사람 나이로 불가능한 값입니다. `orders`의 `quantity` `max`가 **100**이라면 정상 주문으로 보기 어렵죠. 이것이 **이상치(outlier)** 의 신호입니다. (이상치를 *어떻게 처리할지*는 다음 시간의 주제입니다.)

> **읽는 법:** `channel`에 `'web'`, `'app'`뿐 아니라 `'app '`(뒤 공백), `'APP'`(대문자)가 섞여 있습니다. 사람 눈에는 같지만 컴퓨터에는 **다른 값**이라 집계가 어긋납니다. `region`도 `' 서울 '`, `'Seoul'`처럼 표기가 혼재합니다.

진단 결과를 정리하면 다음과 같습니다. 오늘은 **발견**까지가 목표입니다.

| 문제 유형 | 어디서 | 발견 방법 |
| --- | --- | --- |
| 결측치(missing value) | `orders.amount`, `customers.gender` | `isnull().sum()` |
| 이상치(outlier) | `customers.age`(999, 음수), `orders.quantity`(100) | `describe()` |

| 중복 행 | `orders` | `duplicated().sum()` |

## 💡 쉽게 말하면 — 차트는 질문에 답하는 도구

차트 선택의 출발점은 그래프 종류가 아니라 **질문**입니다.

```text
알고 싶은 것          →   차트 유형
한 변수의 분포 모양?       히스토그램 / 박스플롯
두 변수의 관계?            산점도(scatter)
시간에 따른 추이?          선 그래프(line)
범주별 크기 비교?          막대 그래프(bar)
```

> 📌 **실무 포인트:** 차트를 그리기 전에 항상 "이 그림으로 무엇을 판단할 것인가"를 먼저 정합니다. 목적 없는 차트는 예쁘기만 하고 결정을 돕지 못합니다.

## 자세히 알아보기 — 목적 → 차트 매핑

| 알고 싶은 것 | 차트 유형 | matplotlib/seaborn |
| --- | --- | --- |
| 분포(distribution) | 히스토그램, 박스플롯 | `sns.histplot`, `sns.boxplot` |
| 관계(relationship) | 산점도 | `sns.scatterplot` |
| 추이(trend) | 선 그래프 | `plt.plot`, `sns.lineplot` |
| 비교(comparison) | 막대 그래프 | `sns.barplot`, `plt.bar` |

> 💡 차트의 한글 라벨은 환경에 따라 깨질 수 있어, 아래 예제는 **영어 라벨**을 사용합니다. (Part 0에서 한글 폰트를 설정했지만, 안전을 위해 그래프 글자는 영어로 둡니다.)

> **읽는 법:** 히스토그램은 막대가 높은 구간에 값이 몰려 있음을 뜻합니다. 박스플롯의 상자는 가운데 50%, 상자 밖 점들은 **이상치 후보**입니다. 두 그림 모두 "금액이 낮은 쪽에 몰려 있고 높은 금액이 드물게 있다"는 한쪽 쏠림(skew)을 보여줍니다.

> **읽는 법:** 점들이 우상향으로 모이면 양의 관계, 흩어져 있으면 뚜렷한 관계가 없다는 뜻입니다. (나이 999 같은 이상치가 오른쪽 끝에 외따로 찍혀 보일 수 있습니다 — 시각화가 이상치를 잡아내는 또 다른 방법입니다.)

> **읽는 법:** 막대 길이로 범주 간 크기를 한눈에 비교합니다. 여기서 `str.strip()`으로 앞뒤 공백을 임시로 없앴는데, 그래도 `'Seoul'`은 `'서울'`과 따로 세어집니다 — 표기 혼재가 시각화까지 어떻게 번지는지 보여주는 장면입니다.

> 📌 **다른 산업에서는?** 같은 4가지 목적은 어디서나 같습니다. 마케팅은 채널별 전환을 **막대**로, 금융은 주가 **추이**를 선으로, 헬스케어는 검사 수치 **분포**를 히스토그램으로, 제조는 두 공정 변수의 **관계**를 산점도로 봅니다.

## 한 장 정리표

| 주제 | 핵심 한 줄 | 대표 코드 |
| --- | --- | --- |
| 자료형·컬렉션 | 값 하나는 자료형, 여럿은 컬렉션 | `[ ]` `( )` `{ }` `{k:v}` |
| 함수·클래스 | 반복은 함수로, 데이터+작업은 클래스로 | `def f(): return ...` |
| NumPy | 배열 전체에 한 번에(벡터화) | `arr * 1.1`, `arr.mean()` |
| pandas | 섞인 자료형의 표를 다룸 | `df[df["c"] > 0]` |
| 진단 | 계산 전에 진찰부터 | `df.shape / .info() / .describe()` |
| 시각화 | 질문이 차트를 결정 | `sns.histplot / scatterplot` |

````python
# [분포] 히스토그램 + 박스플롯 — 주문 금액은 어떻게 퍼져 있나?
amount_clean = orders["amount"].dropna()   # 결측 제외하고 시각화

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
sns.histplot(amount_clean, bins=20, ax=axes[0], color="#4C72B0")
axes[0].set_title("Distribution of Order Amount (Histogram)")
axes[0].set_xlabel("amount"); axes[0].set_ylabel("count")

sns.boxplot(x=amount_clean, ax=axes[1], color="#55A868")
axes[1].set_title("Order Amount (Boxplot)")
axes[1].set_xlabel("amount")
plt.tight_layout(); plt.show()
````

````python
# [관계] 산점도 — 나이와 주문 금액은 관계가 있을까?
# customers와 orders를 customer_id로 연결(merge는 다음 모듈 주제, 여기선 맛만)
merged = orders.merge(customers[["customer_id", "age"]], on="customer_id", how="left")
sample = merged.dropna(subset=["amount"]).sample(300, random_state=42)

plt.figure(figsize=(8, 5))
sns.scatterplot(data=sample, x="age", y="amount", alpha=0.5)
plt.title("Age vs Order Amount (Scatter)")
plt.xlabel("age"); plt.ylabel("amount")
plt.show()
````

````python
# 공공데이터 스타일 데이터 생성 (따릉이 일별 대여 기록 — 가상)
np.random.seed(7)
n_days = 180
dates = pd.date_range("2025-01-01", periods=n_days, freq="D")
# 계절성: 봄·여름에 대여 증가
seasonal = 3000 + 2500 * np.sin(np.linspace(0, np.pi, n_days))
rentals = (seasonal + np.random.normal(0, 400, n_days)).round().astype(int)

bike = pd.DataFrame({
    "date": dates,
    "rentals": rentals,
    "avg_temp": (np.linspace(-2, 28, n_days) + np.random.normal(0, 2, n_days)).round(1),
    "rainfall": np.random.choice([0, 0, 0, 0, 5, 12, 30], n_days),
    "station_area": np.random.choice(["강남", "마포", "송파", "종로", " 강남 "], n_days),
})
# 오염 심기: 대여수 결측, 기온 이상치, 중복 행
bike.loc[np.random.choice(n_days, 8, replace=False), "rentals"] = np.nan
bike.loc[50, "avg_temp"] = 250.0     # 센서 오류로 보이는 이상치
bike = pd.concat([bike, bike.iloc[[3]]], ignore_index=True)  # 중복 1건

print("따릉이 데이터 준비 완료:", bike.shape)
bike.head()
````

````python
# 시나리오 1 — 구조 진단
print("shape:", bike.shape)
bike.info()
display(bike.describe())
````

````python
# 시나리오 2 — 품질 진단
print("[결측치]")
print(bike.isnull().sum())
print("\n중복 행:", bike.duplicated().sum())
print("\n[station_area 값의 종류]")
print(bike["station_area"].unique())
print("\n기온 이상치 확인(상위 3개):")
print(bike["avg_temp"].sort_values(ascending=False).head(3).tolist())
````

````python
# 시나리오 3 — 시각화 (이상치 제거 후)
clean = bike[(bike["avg_temp"] < 60)].dropna(subset=["rentals"])

fig, axes = plt.subplots(1, 2, figsize=(13, 4))
# 추이
axes[0].plot(clean["date"], clean["rentals"], color="#4C72B0")
axes[0].set_title("Daily Bike Rentals (Trend)")
axes[0].set_xlabel("date"); axes[0].set_ylabel("rentals")
axes[0].tick_params(axis="x", rotation=45)
# 관계
sns.scatterplot(data=clean, x="avg_temp", y="rentals", alpha=0.5, ax=axes[1])
axes[1].set_title("Temperature vs Rentals (Relationship)")
axes[1].set_xlabel("avg_temp"); axes[1].set_ylabel("rentals")
plt.tight_layout(); plt.show()
````
